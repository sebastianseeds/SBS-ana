```python
# Assuming 'file' is your uproot file object and "P" is the tree
p_tree = file["P"]  # Access the tree named "P"

# List all branches in the tree
branches = p_tree.keys()
print(branches)
```

```python
for branch_name in branches:
    branch = p_tree[branch_name]
    # Accessing the first entry in each branch
    # Be cautious with large datasets; you might not want to print everything
    data = branch.array(library="np")[0]  # Using 'library="np"' to get a NumPy array
    print(f"{branch_name}: {data}")
```
